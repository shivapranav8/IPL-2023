```python

#importing necessary libraries
import pandas as pd
import numpy as np
```

```python
df_match = pd.read_csv('/content/IPL_Match_Summary.csv')
df_match.head(10)
```

```python
df_match = df_match.drop(df_match.columns[-1],axis=1)
df_match.head()
```

```python
df_match.rename({'Match No.':'Match_id'}, axis=1, inplace = True)
df_match.head()
```

```python
df_match['Match_id'] = df_match['Match_id'].replace(['Qualifier1', 'Qualifier2'], 'Qualifier')
df_match['Margin'] = df_match['Margin'].fillna('No result')
df_match.tail(30)

```

```python
df_batting = pd.read_csv('/content/IPL_Batting_Scorecard.csv')
df_batting.head(5)
```

```python

df_batting.rename({'Batsman Name':'batsmen', 'selection1_batsmen_Out_NotOut':'out_notout','selection1_batsmen_Runs':'runs', 'selection1_batsmen_Balls':'balls', 'selection1_batsmen_Fours':'fours', 'selection1_batsmen_Sixes':'sixes', 'selection1_batsmen_Strike_Rate':'sr', 'selection1_batsmen_Team':'team', 'selection1_batsmen_Match_No':'Match_id' }, axis=1, inplace = True)
df_batting.head(20)
```

```python
to_drop = df_batting[df_batting['runs'].str.contains('Did not bat:', na=False)].index
df_batting = df_batting.drop(to_drop)
df_batting = df_batting.reset_index(drop=True)
df_batting.head(20)
```

```python
df_batting['out_notout']= np.where(df_batting['out_notout']=='not out',df_batting['out_notout'],'out')
df_batting.head(30)
```

```python
df_batting['Match_id'] = df_batting['Match_id'].str.replace(r'^(Q\w+)\s(.*)', r'\1 \2', regex=True)
df_batting['Match_id'] = df_batting['Match_id'].str.replace(r'^(\w+)\s(.*)', r'\1', regex=True)

df_batting.head(30)
```
